```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import csv

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python

data = []
with open('Market_Basket_Optimisation.csv') as file:
    reader = csv.reader(file, delimiter=',')
    for row in reader:
        data +=[row]
```

```python
data[1:10]
```

```python
len(data)
```

```python
te = TransactionEncoder()
x = te.fit_transform(data)
x
```

```python
te.columns_
```

```python
df = pd.DataFrame(x,columns = te.columns_)
```

```python
df
```

```python
freq_items_set = apriori(df, min_support=0.01, use_colnames=True)
freq_items_set
```

```python
freq_items_set.max()
```

```python
rules = association_rules(freq_items_set, metric='confidence', min_threshold=0.10)
```

```python
rules
```

```python
rules = rules[['antecedents','consequents','support','confidence']]
rules
```

```python
rules[rules['antecedents'] == {'burgers'}] ['consequents']
```
